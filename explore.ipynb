```python
full_text = glob.glob('txts/gutenberg_full_text/*.txt')
```

```python
these_ids = []
for i in full_text:
    ids = ";".join(i.split('/')[-1].split('.')[0].split('_'))
    these_ids.append(ids)
```

```python
for i in meta:
    df = pd.read_csv(i, encoding='latin-1')
    try:
        print(i, len(df.loc[df['gutenberg_id'].isin(these_ids)]))
    except:
        pass
```

```python
df_cather = df.loc[df['source'] == 'lavin']
```

```python
def make_path(row):
    gutenberg_id = row['gutenberg_id']
    this_path = f'txts/other_tsv/{gutenberg_id}.tsv'
    return this_path
df_cather['filepath'] = df_cather.apply(make_path, axis=1)
df_cather
```

```python
full_text = []
for e, row in df_cather.iterrows():
    df = pd.read_csv(row['filepath'], sep='\t', header=None)
    df.columns = ['word', 'count']
    df['source'] = row['gutenberg_id']
    full_text.append(df)
```

```python
novel_text_joined = pd.concat(full_text)
```
